```python
from scipy.stats import poisson, erlang
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```python
mu = 1.3
dist = poisson(mu)
```

```python
xs = np.arange(0, 10)
```

```python
plt.title("Count of obstacles")
plt.xlabel("obstacles")
plt.ylabel("probability")
plt.grid()
plt.xticks(xs)
plt.scatter(xs, [dist.pmf(x) for x in xs]);
```

```python
stat = [dist.rvs() for i in range(1000)]
```

```python
keys, values = zip(*Counter(stat).items())
```

```python
plt.pie(values, labels=keys);
```
